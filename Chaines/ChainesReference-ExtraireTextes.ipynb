```python
import pyperclip,codecs,re
```

```python
from lxml import etree as ET
```

```python
parser = ET.XMLParser(remove_blank_text=True)
```

```python
def extraireTexteTRS(nomTRS):
    xmlTRS=ET.parse(nomTRS,parser)
    lignes=[]
    for t in xmlTRS.xpath("//Turn/text()"):
        lignes.append(t.strip())
    texte="\n".join(lignes)
    return texte    
```
