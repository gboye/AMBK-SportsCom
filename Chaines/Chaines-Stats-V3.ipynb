## Parser les matchs

```python
match1="FrAr"
match2="FrNz"
nFile1=match1+"-RefSynTour.xlsx"
nFile2=match2+"-RefSynTour.xlsx"
```

```python
def renameValues(lDf):
    lDf["type"]=lDf["type"].str.replace("Color","Colour")
    lDf["speaker"]=lDf["speaker"].str.replace("spk1","Journaliste")
    lDf["speaker"]=lDf["speaker"].str.replace("spk2","Expert")
    lDf["Mi-temps"]=lDf["Mi-temps"].fillna("")
    lDf["Mi-temps"]=lDf["Mi-temps"].replace(1,"MT1")
    lDf["Mi-temps"]=lDf["Mi-temps"].replace(2,"MT2")
    return lDf    
```

# Longueur des chaines

## Préparer les tableaux
- groupby "Mi-temps","speaker","type","numChaine" => longueur=chaineIndex.max()
- groupby "Mi-temps","speaker","type","chaineIndex" => nb=numChaine.count()
- pivot index="longueur", columns=\["speaker","type","mi-temps"\], values="nb"

```python
def getLongMatch(nFile):
    iDf=pd.read_excel(rep+nFile,index_col=0)
    df=renameValues(iDf)
    
    selection=(df.speaker.isin(["Journaliste",
                                "Expert"]
                              ))&(df.type.isin(["Colour commentary",
                                                "Play by play"]))

    dfExtract0=df[selection].groupby(["Mi-temps",
                                      "speaker",
                                      "type",
                                      "numChaine"]
                                    )[["chaineIndex"]].max()

    dfExtract0=dfExtract0.reset_index()
    dfExtract=dfExtract0.groupby(["Mi-temps",
                                  "speaker",
                                  "type",
                                  "chaineIndex"]
                                )[["numChaine"]].count().reset_index()
    dfExtract.columns="mi-temps speaker type longueur nb".split(" ")
    result=dfExtract.pivot(index="longueur",
                           columns=["speaker","type","mi-temps"],
                           values="nb").fillna(0)
    return result
```

### Parser les deux matchs

```python
df1=getLongMatch(nFile1)
df2=getLongMatch(nFile2)
```

### Assembler le tableaux avec les 2 matchs
- axis=1 pour mettre les colonnes à la suite
- keys=\[match1,match2\] pour ajouter un niveau d'index dans les colonnes
- names=\["match","speaker","type","mitemps"\] pour donner un nom parlant à chaque niveau
- astype(int) pour éviter les nombres à virgules pour les nb

```python
df=pd.concat([df1,df2],axis=1,keys=[match1,match2],names=["match","speaker","type","mitemps"]).fillna(0)
df.columns=df.columns.reorder_levels([0,3,2,1])
df.index=df.index.astype(int)
```

```python
df.to_excel(writer, sheet_name="Dist-Long")
maxLChaines=df.index.max()
```

### Calculer les longueurs moyennes

```python
def calculMoyPond(col):
    prodPond=0
    somme=float(col.sum())
    for i in range(maxLChaines):
        n=i+1
        prodPond+=col.loc[n]*n
    return round(prodPond/somme,1)
```

```python
selections=[["type"],["speaker"],["match"],["type","speaker"],["match","type"],["match","speaker"],["match","mitemps"],["match","mitemps","speaker"],["match","mitemps","type"],["match","mitemps","type","speaker"]]
moyDict={}
for selection in selections:
    selectionName="-".join(selection)
    moyDict=moyDict|df.T.groupby(by=selection).sum().T.apply(calculMoyPond).to_dict()
```

```python
pd.DataFrame.from_dict(moyDict,orient="index",columns=["Longueur moyenne"]).to_excel(writer, sheet_name="Long-Moyennes")
```

## Fabriquer les graphes

Pour extraire une sous-table :
- on donne la sélection et le niveau où elle se situe 
> df.xs(selection, level=selection_level, axis=1, drop_level=True)

```python
def makeDfGraphe(lDf,cut=4):
    dfGraph=lDf.iloc[:cut].copy()
    autres=pd.DataFrame(lDf.iloc[cut:].sum()).T
    autres.index=["%d+"%(cut+1)]
    dfGraph=pd.concat([dfGraph,autres])
    return dfGraph
```

```python
dfCut=makeDfGraphe(df)
dfCut.columns.tolist()
dfCut=dfCut[sorted(dfCut.columns.tolist(), key=lambda tup: (tup[0],tup[1],tup[2],tup[3]) )]
dfCut
```

```python
def saveGraphe(lDf,name,figsize=(10,30),key="Long"):
    # display(lDf)
    nSubplots=len(lDf.columns)
    nPairs=round(nSubplots/2.0)
    print (nSubplots,nPairs)
    lDf.plot.pie(colors=colors,autopct='%.0f%%',figsize=figsize,
                 # title="Distribution des longueurs",
                 legend=False,subplots=True,layout=(nPairs,2))
    print("JE-PC-%sChaines-%s.png"%(key,name))
    lDf.to_excel(writer, sheet_name=key+name)
    plt.savefig("JE-PC-%sChaines-%s.png"%(key,name),
                bbox_inches="tight",dpi=300)
    return
```

# Traitement des roots

```python
def getRootMatch(nFile):
    cats="PRO PRO:PER PRO:REL PRO:DEM PRO:IND PRO:POS NAM NOM DET:POS".split(" ")
    iDf=pd.read_excel(rep+nFile,index_col=0)
    df=renameValues(iDf)
    
    selection=(df.speaker.isin(["Journaliste","Expert"]))\
            & (df.type.isin(["Colour commentary","Play by play"])\
            & (df.chaineDep==0)\
            & (df.Cat.isin(cats)))

    dfExtract=df[selection].groupby(["Mi-temps",
                                      "speaker",
                                      "type",
                                      "Cat"]
                                    )[["numChaine"]].count()

    dfExtract=dfExtract.reset_index()
    dfExtract.columns="mi-temps speaker type cat nb".split(" ")
    result=dfExtract.pivot(index="cat",
                           columns=["speaker","type","mi-temps"],
                           values="nb").fillna(0)
    return result
```
