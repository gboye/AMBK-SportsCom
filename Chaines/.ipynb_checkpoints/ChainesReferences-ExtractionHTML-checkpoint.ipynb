```python
def lireChainesXL(fNom):
    dfBrut=pd.read_excel(fNom, sheet_name="Sheet1")
    dfResult=dfBrut[[ \
                         "Transcription", "Cat", "Lemme", \
                         "Referent", \
                         "numChaine", "chaineIndex", "chaineDep", "chaineApp", \
                        ]]    
    dfResult.loc[:,"Referent"]=dfResult.Referent.str.strip()
    dfResult.loc[:,"numChaine"]=dfResult.numChaine.astype("Int64",errors="ignore")
    dfResult.loc[:,"chaineIndex"]=dfResult.chaineIndex.astype("Int64",errors="ignore")
    dfResult.loc[:,"chaineDep"]=dfResult.chaineDep.astype("Int64",errors="ignore")
    dfResult.loc[:,"Transcription"]=dfResult.Transcription.astype("unicode")
#     display(dfResult)
    return dfResult
    
```

# Formatage du texte

```python
sub='<sub style="font-size:8px;position:relative;">%s</sub>'
sup='<sup style="font-size:8px;position:relative;">%s</sup>'
```

```python
texte=[]
phrase=[]

def formatPhrase(row):
    if row.chaineIndex==row.chaineIndex:
        rIndex=sub%(row.chaineIndex)
    else:
        rIndex=sub%"?"
        
    nom=row.Transcription
    
    if row.chaineDep==row.chaineDep:
        rDep=sub%(row.chaineDep)
    else:
        rDep=sub%"?"

    if row.Referent==row.Referent:
        rRef=sup%(row.Referent)
    else:
        rRef=sup%"?"

    result='''<math display="inline">
  <mmultiscripts>
    <mi>%s</mi> <!-- base expression -->
    <mi style="font-size:8px;">%s</mi>
    <mi style="font-size:8px;">%s</mi>
    <mprescripts />
    <mi style="font-size:8px;">%s</mi>
    <mi></mi>
  </mmultiscripts>
</math>'''%(nom,rDep,rRef,rIndex)
    
    return result

for idx,row in dfChaines.iterrows():
    if row.Referent==row.Referent:
        # phrase.append(sub%(row.chaineIndex)+
        #               row.Transcription+
        #               sub%(row.chaineDep)+
        #               sup%(row.Referent)
        #              )
        phrase.append(formatPhrase(row))
    elif row.Cat=="SENT":
        phrase.append(row.Transcription)
        texte.append(" ".join(phrase))
        phrase=[]
    elif row.Transcription!="PUN;cit":
        phrase.append(row.Transcription)
```

```python
with codecs.open(fHTML,"w",encoding="utf8") as outFile:
    for t in texte:
        t=t.replace(" ,",",").replace(" .",".").replace("' ","'")
        outFile.write(t+"<br/><br/><br/>\n\n")
```
