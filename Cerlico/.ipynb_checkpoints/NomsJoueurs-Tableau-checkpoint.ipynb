```python
import re,codecs
from lxml import etree
import pandas as pd
```

```python
fJoueurs="NomsJoueurs.txt"
with codecs.open(fJoueurs,"r",encoding="utf8") as inFile:
    lJoueurs="".join(inFile.readlines())
```

```python
tEquipes=re.split(u"\n(?=.*:)",lJoueurs)
```

```python
equipes={}
for tEquipe in tEquipes[1:]:
    print tEquipe
    lEquipe=tEquipe.split("\n")
    nomEquipe=lEquipe[0].split(":")[0].strip()
    equipes[nomEquipe]={}
    equipe=equipes[nomEquipe]
    for ligne in lEquipe[1:]:
        m=re.match(ur"(.*)\t(.*)",ligne)
        if m:
            print "poste",m.group(1)
            print "joueurs",m.group(2).split(",")
            poste=m.group(1).strip()
            joueurs=[j.strip() for j in m.group(2).split(",") if j.strip()!=""]
            if joueurs:
                equipe[poste]=joueurs
    
```

```python
def getJoueurs(pays):
    result=[]
    for k,v in equipes[pays].iteritems():
        result.extend(v)
        result=[j.replace("(cap.)","").strip() for j in result]
    return result
```

```python
joueursFr=getJoueurs("France")
joueursAr=getJoueurs("Argentine")
```

```python
prenomsFr=set()
nomsFr=set()
for j in joueursFr:
    prenom,nom=j.split(" ",1)
    prenomsFr.add(prenom)
    nomsFr.add(nom)
prenomsFr,nomsFr   
```

```python
prenomsAr=set()
nomsAr=set()
for j in joueursAr:
    prenom,nom=j.split(" ",1)
    prenomsAr.add(prenom)
    nomsAr.add(nom)
prenomsAr,nomsAr   
```

```python
coordJoueursFr=u"|".join(prenomsFr|nomsFr)
coordJoueursAr=u"|".join(prenomsAr|nomsAr)
regexJoueurs=u"(%s|%s)"%(coordJoueursFr,coordJoueursAr)
```

```python
repRugby="/Volumes/BroadDisk/Rugby/"
fichierAssociations="Videos/CorrespondancesAction-PlayVsColourVsSupportaire.csv"
dfAsso=pd.read_csv(repRugby+fichierAssociations,sep=";",keep_default_na=False)
```

```python
dfAsso.columns=[u'num', u'TimeDeb', u'Fin', u'penalites', u'Combinaisons',
       u'Evenements', u'Actions', u'CommNum', u'CommText', u'Erreurs1',
       u'Erreurs2', u'Erreurs3', u'Erreurs4', u'CommType',
       u'CommTypeCorr']
```

```python
dfAsso["joueurFr"]=dfAsso[dfAsso["CommText"].str.contains(coordJoueursFr)]["CommText"].str.split(" ").str[0]
dfAsso["joueurAr"]=dfAsso[dfAsso["CommText"].str.contains(coordJoueursAr)]["CommText"].str.split(" ").str[0]
```

```python
tourFr=dfAsso[dfAsso["joueurFr"].notnull()]["joueurFr"].tolist()
tourAr=dfAsso[dfAsso["joueurAr"].notnull()]["joueurAr"].tolist()
```

```python
tourActionsFr={}
for tour in tourFr:
    t=tour.lower()
    actions=(dfAsso[dfAsso["CommNum"].str.lower().str.contains(t)]["Actions"]).tolist()
    if actions:
        tourActionsFr[t]=actions
```

```python
tourActionsAr={}
for tour in tourAr:
    t=tour.lower()
    actions=(dfAsso[dfAsso["CommNum"].str.lower().str.contains(t)]["Actions"]).tolist()
    if actions:
        tourActionsAr[t]=actions
```
