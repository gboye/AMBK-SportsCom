```python
import pandas as pd
import re,yaml
from cellbell import ding
```

```python
testPerception="frFiltre"
nbQuestions=21
```

# Lire le fichier excel en entrée

```python
dfDonneesBrutes=pd.read_excel("Reponses.xlsx")
```

# Repérer les questions

```python
colonnes=["Question %d"%(i+1) for i in range(nbQuestions)]
colonnes
```

```python
dfQuestions=dfDonneesBrutes[colonnes]
```

## Repérer les liens des stimulis pour les numéros de question du quizz

```python
lienStimuli={}
dictQuestions=dfQuestions.drop_duplicates().to_dict()
for k,v in dictQuestions.iteritems():
    print k
    numQuestion=k.split(" ")[-1]
    for k1,v1 in v.iteritems():
        m=re.match(ur".*(http\S+)\s.*",v1)
        lienStimuli[numQuestion]=m.group(1).strip()
lienStimuli
```

## Repérer les liens des stimulis sur Humanum pour chaque découpage

```python
fCorrespondances="EmotionsLiensSons.yaml"
with open(fCorrespondances,"r") as inFile:
    dictCorrespondances=yaml.load(inFile)
dictCorrespondances#["Humanum"]["fr"]
```

```python
dictLien={v:k for k,v in dictCorrespondances["Humanum"][testPerception].iteritems()}
```

## Faire la correspondances entre les questions et les stimulis

```python
dictStimuli={}
for n in lienStimuli:
    dictStimuli[n]=dictLien[lienStimuli[n]]
dictReponses={}
for i in range(nbQuestions):
    num=i+1
    numStimuli=dictStimuli[str(num)]
    print num, dictStimuli[str(num)], lienStimuli[str(num)]
    dictReponses[numStimuli]=num
dictReponses
```

### Récupérer les métadonnées des stimulis

```python
dfMetaDonnees=pd.read_excel("ActionsJaponFidji.xlsx")
dfMetaDonnees["nStimuli"]=dfMetaDonnees.index+1
dfMetaDonnees["nStimuli"]=dfMetaDonnees["nStimuli"]
dfMetaDonnees["Stimuli"]=u"Stimuli "+dfMetaDonnees["nStimuli"].apply(str)
dfMetaDonnees[u"URL"]=dfMetaDonnees["nStimuli"].map(dictCorrespondances["Humanum"][testPerception])
dfMetaDonnees[u"nRéponse"]=dfMetaDonnees["nStimuli"].map(dictReponses)
dfMetaDonnees[u"Réponse"]=u"Réponse "+dfMetaDonnees[u"nRéponse"].apply(str)
colonnes=u"Stimuli Réponse URL Action Minutes Avantage".split(" ")
dfMetaDonnees=dfMetaDonnees[colonnes]
dfMetaDonnees
```

# Sélectionner les réponses

```python
colonnes=["Nom",u"Prénom"]
# colonnes+=["Question %d"%(i+1) for i in range(21)]
colonnes+=[u"Réponse %d"%(i+1) for i in range(nbQuestions)]
colonnes
```

```python
dfDonnees=dfDonneesBrutes[colonnes]
```

# Transformer la matrice pour obtenir une question par ligne

```python
dfReponses=dfDonnees.melt(id_vars=["Nom",u"Prénom"])
dfReponses=dfReponses[dfReponses["value"]!="-"].reindex()
```

# Extraire les valeurs des réponses des champs Réponses

```python
def getReponse(row,i):
    lReponses=row["value"].split(";")
    lValeur=lReponses[i-1].split(": ")[-1]
    if lValeur.strip()!="":
        result=int(eval(lValeur))
    else:
        result=None
    return result
```

```python
for i in range(1,4):
    dfReponses["P%d"%i]=dfReponses.apply(lambda x: getReponse(x,i),axis=1)
```

```python
dfReponses["1"]=1
```

```python
dfReponses
```

# Transformer la table des réponses par valeur en colonne par valeur

```python
tableP1=pd.pivot_table(dfReponses[["Nom",u"Prénom","variable","P1","1"]],
               values=["1"],
              index=["Nom",u"Prénom","variable"], 
              columns=["P1"]).reset_index()
```

```python
tableP2=pd.pivot_table(dfReponses[["Nom",u"Prénom","variable","P2","1"]],
               values=["1"],
              index=["Nom",u"Prénom","variable"], 
              columns=["P2"]).reset_index()
```

```python
tableP3=pd.pivot_table(dfReponses[["Nom",u"Prénom","variable","P3","1"]],
               values=["1"],
              index=["Nom",u"Prénom","variable"], 
              columns=["P3"]).reset_index()
```

# Assembler les tables

## applatir les index et différencier les colonnes par partie

```python
tableP1.columns=u"Nom Prénom Réponse P1-1 P1-2 P1-3 P1-4 P1-5".split(" ")
#tableP1
```

```python
tableP2.columns=u"Nom Prénom Réponse P2-1 P2-2 P2-3 P2-4 P2-5".split(" ")
#tableP2
```

```python
tableP3.columns=u"Nom Prénom Réponse P3-1 P3-2 P3-3 P3-4 P3-5".split(" ")
#tableP3
```

## Fusionner les parties en un seul tableau

```python
tables=pd.merge(tableP1,tableP2)
tables=pd.merge(tables,tableP3)
tables
```

```python
tableauFinal=pd.merge(tables,dfMetaDonnees)
tableauFinal.to_excel("TableauFinal-%s.xlsx"%testPerception,sheet_name=testPerception)
```
