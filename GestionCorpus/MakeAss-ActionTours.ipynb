### Faire une liste de tuples à partir d'un df
https://stackoverflow.com/questions/45285244/convert-a-dataframe-to-list-of-tuples

```python
tTours=[tuple(r) for r in dfTours["startTime endTime speaker tour".split(" ")].to_numpy()]
tActions=[tuple(r) for r in dfActions["startTime endTime action".split(" ")].to_numpy()]
```

```python
'''
Dialogue: 0,0:12:24.94,0:12:29.30,action,,0,0,0,,A-001 grand coup de pied
Dialogue: 0,0:12:28.56,0:12:30.92,spk1,spk1,0,0,0,,spk1-002  est donné par David Skrela.
'''
```

```python
def time2ass(chaine):
    result=chaine[1:-4]
    return result
```

```python
lTours=[]
for nT,(sT,eT,spk,tour) in enumerate(tTours):
    if " " in spk:
        style="pb"
    else:
        style=spk
    if tour==tour:
        tour=tour.replace("\n","\\n")
    else:
        tour=""
    lTours.append("Dialogue: 0,%s,%s,%s,%s,0,0,0,,Tour-%03d (%s)\\n %s"%(time2ass(sT),time2ass(eT),style,spk,nT+1,spk,tour))
```

```python
lTours
```

```python
lActions=[]
for nA,(sA,eA,action) in enumerate(tActions):
    lTours.append("Dialogue: 0,%s,%s,action,,0,0,0,,Action-%03d %s"%(time2ass(sA),time2ass(eA),nA+1,action))
```

```python
lAss=sorted(lTours+lActions)
print "\n".join(lAss)
pyperclip.copy("\n".join(lAss))
```
