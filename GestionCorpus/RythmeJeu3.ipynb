```python
timeMT1="00:11:52.40"
timeMT2="00:57:22.70"
debutMT2=325760.8
decalage1=0
decalage2=600000
def decodeTime(temps):
    (heure,minute,seconde)=temps.split(":")
    timeCode=int(100*(3600*int(heure)+60*int(minute)+float(seconde))-decalage1)
    if timeCode>=decalage2:
        timeCode=timeCode-decalage2+int(debutMT2)
    return timeCode

def decodeTimecode(temps):
    (heure,minute,seconde)=temps.split(":")
    timeCode=int(100*(3600*int(heure)+60*int(minute)+float(seconde)))
    return timeCode

def codeTimecode(timecode):
    secondes=float(timecode)/100
    heures=secondes//3600
    secondes=secondes-heures*3600
    minutes=secondes//60
    secondes=secondes-minutes*60
    return "%02d:%02d:%05.2f"%(heures,minutes,secondes)

decalage1=decodeTime(timeMT1)
decalage2=decodeTime(timeMT2)
print decalage1, decodeTime(timeMT1)
print decalage2, decodeTime(timeMT2)
```

```python
def adapte(valeur):
    return float(valeur*100)
```

# Attention dans cette version le rythme est calculé d'après le jeu à T

```python
%%time
rythmes={}
vitesses={}
#Tous les rythmes sont ramenés à 5 secondes comme pour Berlin
echelle=5
demiFenetres=[1,2,3,4,5,6,7,8]
increment=min(min(demiFenetres)*2,1)
for demiFenetre in demiFenetres:
    rythmes[demiFenetre]={}
    vitesses[demiFenetre]={}
    demiFen=adapte(demiFenetre)
    debut=adapte(30)
    fin=adapte(6300)
    step=adapte(increment)
    for i in arange(debut,fin,step):
#        nbActions=len(timeLine[(timeLine.index<=i+demiFen) & (timeLine.index>=i-demiFen)])
        nbActions=len(timeLine[(timeLine.index<=i) & (timeLine.index>=i-2*demiFen)])
        rythmes[demiFenetre][float(i)/100]=float(nbActions)*100/(2*demiFen)*echelle
#        nbActionsCommentees=len(timeLine[(timeLine.index<=i+demiFen) & (timeLine.index>=i-demiFen) & (timeLine["comment"]!="")])
        nbActionsCommentees=len(timeLine[(timeLine.index<=i) & (timeLine.index>=i-2*demiFen) & (timeLine["comment"]!="")])
        vitesses[demiFenetre][float(i)/100]=float(nbActionsCommentees)*100/(2*demiFen)*echelle
        
```

```python
def calculerRythme(rythmeCalc,timelineRef):
    resultats={}
    elementCle=timelineRef.keys()[0]
    for elementTime in sorted(timelineRef[elementCle]):
        if elementTime < 6300:
            ligneResultat=[]
            for rythme in sorted(timelineRef):
                ligneResultat.append(timelineRef[rythme][elementTime])
            resultats[elementTime]=ligneResultat
            rythmeCalc.loc[int(adapte(elementTime))]=ligneResultat
```

```python
%%time
timeRythmes=pd.DataFrame(columns=["RJ%.2f"%i for i in demiFenetres])
calculerRythme(timeRythmes,rythmes)
#timeRythmes
```
