```python
def makeTimeline(tableau,colName="timeline"):
    if not "timeline" in tableau.columns and not "timecode" in tableau.columns:
        tableau=tableau.reset_index()
        tableau[colName]=tableau["index"]
        del tableau["index"]
    return tableau

def shiftTimeline(tableau):
    if "timecode" in tableau.columns:
        tableau["timeline"]=tableau["timecode"].dropna()
        tableau.loc[tableau["timecode"]>decalageVideoTRS["timecode"],"timeline"]=tableau["timecode"]+decalageVideoTRS["decalage"]
    return tableau

def codeTimecode(timecode):
    secondes=float(timecode)/100
    heures=secondes//3600
    secondes=secondes-heures*3600
    minutes=secondes//60
    secondes=secondes-minutes*60
    return "%02d:%02d:%05.2f"%(heures,minutes,secondes)
```
