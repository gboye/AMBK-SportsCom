## Corrélation Rythmes Débits MT1

```python
lignesTable=timeLine=pd.DataFrame(columns=["type","rythme","speaker","débit","indice","pvaleur"])
```

```python
from scipy.stats import ttest_ind, pearsonr
#finRythme=2700
#echantillon=timeRythmeDebitVitesse[:finRythme]
echantillon=timeRythmeDebitVitesse[(timeRythmeDebitVitesse["timeline"]<273300)]
numLigne=0
for play in ["PxP","colour","support"]:
    for rythmeJeu in ["RJ%d.00"%i for i in range(1,9)]+["RC%d.00"%i for i in range(1,9)]:
        print play,"\t",rythmeJeu,
        for speaker in ["S1","S2"]:
            rythmesSpeakers=[speaker+"D%d.00"%i for i in range(1,13)]#+["S2D%d.00"%i for i in range(12,13)]
            rythmesSpeakersFill=[speaker+"D%02d.00"%i for i in range(1,13)]#+["S2D%d.00"%i for i in range(12,13)]
    #        rythmesSpeakers+=["S1V%d.00"%i for i in range(1,9)]+["S2V%d.00"%i for i in range(1,9)]
            for numRythme,rythmeSpeaker in enumerate(rythmesSpeakers):
                Base = echantillon[echantillon[play]>0][rythmeSpeaker]
                Correlat=rythmeJeu
                Comparaison = echantillon[echantillon[play]>0][Correlat]
                (indice, pvaleur)=pearsonr(Comparaison,Base)
                lignesTable.loc[numLigne]=[play,rythmeJeu,speaker,rythmesSpeakersFill[numRythme],round(indice,2), pvaleur]
                numLigne+=1

lignesTable=lignesTable.sort_index()
correlationTable=pd.pivot_table(lignesTable,values="indice",columns="rythme",index=["type","speaker","débit"])
correlationTable.to_csv(tableMT1Name,sep=";",encoding="utf8")
```
