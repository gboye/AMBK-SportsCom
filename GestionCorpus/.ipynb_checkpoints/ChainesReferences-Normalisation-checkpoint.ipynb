```python
numChainSupp=10001

def addChains(row):
    global numChainSupp
    if pd.notnull(row["Modif"]):
        modif=row["Modif"]
#         print modif
        m=re.match(ur"C\s*[+=]\s*((\w|-)+)",modif)
        if m:
            newRef=m.group(1)
            print newRef,modif
            row["Referent"]=newRef
        m=re.match(ur"(NC|NR)(.*)",modif)
        if m:
            mType=m.group(1)
            mModif=m.group(2)
            if mType=="NC" or mType=="NR":
#                 print mType
                if pd.isna(row["numChaine"]) or row["numChaine"]<10000:
                    row["numChaine"]=numChainSupp
                    numChainSupp+=1
                if mModif:
#                     print mModif
                    mRef=re.match(ur"(=|\+)(.*)",mModif)
                    if mRef:
                        newRef=mRef.group(2)
                        row["Referent"]=newRef
#                 print
#             elif mType=="NR":
# #                 print mType
#                 if pd.isna(row["numChaine"]) or row["numChaine"]<10000:
#                     row["numChaine"]=numChainSupp
#                     numChainSupp+=1
#                 if mModif:
# #                     print mModif
#                     mRef=re.match(ur"(=|\+)(.*)",mModif)
#                     if mRef:
#                         newRef=mRef.group(2)
#                         row["Referent"]=newRef
    return row
```

```python
dfChaines=dfChaines.apply(addChains,axis=1)
dfChaines[dfChaines["Modif"]=="NC"]
```

## Traitement des modifications de chaînes

```python
dfChaines.oldNumChaine=dfChaines[dfChaines.Referent.notnull()].oldNumChaine.ffill()
```

```python
def updateRows(row):
    numChaine=np.NaN
    chaineIndex=np.NaN
    chaineDep=np.NaN
    idx=row.index
    if pd.isna(row["Modif"]):
        row["numChaine"]=row["oldNumChaine"]
        row["chaineIndex"]=row["oldChaineIndex"]
        row["chaineDep"]=row["oldChaineDep"]
    else:
        modif=row["Modif"]
        m=re.match(ur"R\s*\+\s*(\d+)\s*",modif)
        if row["Modif"]=="A":
            row["numChaine"]=np.NaN
            row["chaineApp"]=row["oldNumChaine"]
        elif row["Modif"]=="D":
            row["Referent"]=np.NaN
        elif row["Modif"]=="IND":
            if pd.isna(row["Referent"]):
                oldRef=""
            else:
                oldRef=str(row["Referent"])
            row["Referent"]="IND "+oldRef
        elif m:
            row["newNumChaine"]=int(m.group(1))            
            row["newChaineIndex"]=row["chaineIndex"]
            row["newChaineDep"]=row["chaineDep"]
            if pd.isna(row["Referent"]):
                row["Referent"]="REPORT"
        else:
            row["numChaine"]=row["oldNumChaine"]
            row["chaineIndex"]=row["oldChaineIndex"]
            row["chaineDep"]=row["oldChaineDep"]            
    return row
```

```python
dfResult=dfChaines.apply(updateRows,axis=1)
```

```python
dfResult.to_excel(fChaines.replace(".xls","-GB.xls"),index=None)
```
