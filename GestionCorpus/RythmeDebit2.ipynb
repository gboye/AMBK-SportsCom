```python
repertoire="/Users/gilles/Copy/Recherche/Rugby/CorpusRugby2007/"
rythmeName="/Users/gilles/Copy/Recherche/Rugby/CorpusRugby2007/Timeline-Rythme-MT1-Pandas.pkl"
debitName="/Users/gilles/Copy/Recherche/Rugby/CorpusRugby2007/Timeline-Debit-Pandas.pkl"
vitesseName="/Users/gilles/Copy/Recherche/Rugby/CorpusRugby2007/Timeline-Vitesse-Pandas.pkl"
commentairesName="/Users/gilles/Copy/Recherche/Rugby/CorpusRugby2007/Timeline-TypeCommentaires.pkl"
commentaireActionName="/Users/gilles/Copy/Recherche/Rugby/CorpusRugby2007/Timeline-CommentaireAction.pkl"
exportName="/Users/gilles/Copy/Recherche/Rugby/CorpusRugby2007/Timeline-RDVTAC.csv"
```

```python
with codecs.open(rythmeName,"rb") as input:
    timeRythmes=pickle.load(input)
```

```python
with codecs.open(debitName,"rb") as input:
    timeDebits=pickle.load(input)
with codecs.open(vitesseName,"rb") as input:
    timeVitesses=pickle.load(input)
with codecs.open(commentairesName,"rb") as input:
    timeCommentaires=pickle.load(input)
```

```python
with codecs.open(commentaireActionName,"rb") as input:
    timeCommentaireAction=pickle.load(input)
```

```python
timeRythmeDebitVitesse=pd.merge(timeRythmes,timeDebits["spk1"],right_index=True,left_index=True)
timeRythmeDebitVitesse=pd.merge(timeRythmeDebitVitesse,timeDebits["spk2"],right_index=True,left_index=True)
timeRythmeDebitVitesse=pd.merge(timeRythmeDebitVitesse,timeDebits["spk3"],right_index=True,left_index=True)
timeRythmeDebitVitesse=pd.merge(timeRythmeDebitVitesse,timeDebits["all"],right_index=True,left_index=True)
timeRythmeDebitVitesse=pd.merge(timeRythmeDebitVitesse,timeVitesses["spk1"],right_index=True,left_index=True)
timeRythmeDebitVitesse=pd.merge(timeRythmeDebitVitesse,timeVitesses["spk2"],right_index=True,left_index=True)
timeRythmeDebitVitesse=pd.merge(timeRythmeDebitVitesse,timeVitesses["spk3"],right_index=True,left_index=True)
timeRythmeDebitVitesse=pd.merge(timeRythmeDebitVitesse,timeVitesses["all"],right_index=True,left_index=True)
timeRythmeDebitVitesse=pd.merge(timeRythmeDebitVitesse,timeCommentaires,right_index=True,left_index=True)
```
