```python
repertoire="/Users/gilles/Copy/Recherche/Rugby/CorpusRugby2007/"
textgridName="/Users/gilles/Copy/Recherche/Rugby/CorpusRugby2007/phonTimelineSorted.csv"
debitName="/Users/gilles/Copy/Recherche/Rugby/CorpusRugby2007/Timeline-Debit.csv"
vitesseName="/Users/gilles/Copy/Recherche/Rugby/CorpusRugby2007/Timeline-Vitesse.csv"
textgridFile=open(textgridName,"r")
textgrid=textgridFile.readlines()
```

```python
voyelles=["a","e","i","y","u","eh","oh","eu","on","an","swa"]
```

```python
timing=[]
for ligne in textgrid[1:]:
    timing.append(ligne.strip().split(","))
```

```python
#timeLine=pd.DataFrame(index=[float(i)/100 for i in range(6500000)],columns=["début","fin","V1","V2","V3"])
timeLine=pd.DataFrame(columns=["début","fin","V1","V2","V3","S1","S2","S3"])
```

```python
%%time
for ligne in timing:
    debut=int(float(ligne[0])*100)
#    debut=float(ligne[0])*100//20*20/100
#    debut=int(float(ligne[0]))
#    print debut,
    if ligne[2] in voyelles:
        if ligne[3]=="spk1":
            timeLine.loc[debut]=[ligne[0],ligne[1],ligne[2],None,None,None,None,None]
        elif ligne[3]=="spk2":
            timeLine.loc[debut]=[ligne[0],ligne[1],None,ligne[2],None,None,None,None]
        elif ligne[3]=="spk3":
            timeLine.loc[debut]=[ligne[0],ligne[1],None,None,ligne[2],None,None,None]
    elif ligne[2]=="SIL":
        if ligne[3]=="spk1":
            timeLine.loc[debut]=[ligne[0],ligne[1],None,None,None,"SIL",None,None]
        elif ligne[3]=="spk2":
            timeLine.loc[debut]=[ligne[0],ligne[1],None,None,None,None,"SIL",None]
        elif ligne[3]=="spk3":
            timeLine.loc[debut]=[ligne[0],ligne[1],None,None,None,None,None,"SIL"]
        
```

```python
#print len(timeLine[timeLine.index<4400][timeLine[timeLine.index<4400].index>3300][["V1","V2","V3"]].dropna(thresh=1))
timeLine["durée"]=timeLine["fin"].astype(float)-timeLine["début"].astype(float)
```

```python
def sliceDataFrame(df,debut,fin,seuil,colonnes):
    return df[df.index<fin][df[df.index<fin].index>debut][colonnes].dropna(thresh=seuil)

silence={"all":["S1","S2","S3","durée"],"spk1":["S1","V2","V3","durée"],"spk2":["S2","V1","V3","durée"],"spk3":["S3","V1","V2","durée"]}
parole={"all":["V1","V2","V3"],"spk1":["V1"],"spk2":["V2"],"spk3":["V3"]}
speakers=parole.keys()
```

```python
#sliceDataFrame(timeLine,3300,4400,1,parole["spk1"])
```

```python
def adapte(valeur):
    return float(valeur*100)
```

```python
demiFenetres=[1,2,3,4,5,6,7,8]
dfDebits={spk:pd.DataFrame(columns=["D%.2f"%i for i in demiFenetres]) for spk in speakers}
dfVitesses={spk:pd.DataFrame(columns=["V%.2f"%i for i in demiFenetres]) for spk in speakers}


increment=min(min(demiFenetres)*2,1)
debut=adapte(30)
fin=adapte(6440)
step=adapte(increment)
ligneVide=[np.nan for i in demiFenetres]
```

```python
#%%time
for i in arange(debut,fin,step):
    for demiFenetre in demiFenetres:
        demiFen=adapte(demiFenetre) 
        for speaker in speakers:
            nbVoyelles=len(sliceDataFrame(timeLine,i-demiFen,i+demiFen,1,parole[speaker]))
            numeroIndex=int(i)
            if not numeroIndex in dfDebits[speaker].index:
                dfDebits[speaker].loc[numeroIndex]=ligneVide
                dfVitesses[speaker].loc[numeroIndex]=ligneVide
            dfDebits[speaker].loc[numeroIndex]["D%.2f"%demiFenetre]=float(nbVoyelles)/(2*demiFen/100)
            sSilence=sliceDataFrame(timeLine,i-demiFen,i+demiFen,2,silence[speaker])["durée"].sum()
            dfVitesses[speaker].loc[numeroIndex]["V%.2f"%demiFenetre]=float(nbVoyelles)/(2*demiFen/100-sSilence)
```
