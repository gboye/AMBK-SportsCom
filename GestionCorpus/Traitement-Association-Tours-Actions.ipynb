```python
import pandas as pd
import numpy as np
```

```python
nAssociations="/Users/gilles/Downloads/NZ-TablesActionsTours2-GB.xlsx"
xls = pd.ExcelFile(nAssociations)
xls.sheet_names
```

```python
dfActions = pd.read_excel(xls, sheet_name=u'Actions',encoding="utf8")
dfActions.columns="nAction startTime endTime action startSec endSec nTour".split(" ")
dfActionsTours=dfActions[dfActions["nTour"].notnull()].copy()
```

```python
dfActionsTours[["nAction","nTour"]]=dfActionsTours[["nAction","nTour"]].astype(str)
# dfActionsTours["nTour"].values
```

```python
dfTours = pd.read_excel(xls, sheet_name=u'Tours',encoding="utf8")
dfTours.columns="nTour speaker startTime endTime tour startSec endSec".split(" ")
dfTours["nTour"]=dfTours["nTour"].astype(str)
# dfTours["nTour"].values
```

```python
dfTours["nActions"]=""
dfTours["PxP"]=""
dfTours["Color"]=""
dfTours["Supporter"]=""
for index,row in dfActionsTours["nAction nTour".split(" ")].iterrows():
    nAction=row["nAction"]
    lTours=row["nTour"].split(";")
    for lTour in lTours:
        iTour=int(lTour)-1
        rowTour=dfTours.at[iTour,"nActions"]
        if rowTour=="":
            dfTours.at[iTour,"nActions"]=nAction
        else:
            dfTours.at[iTour,"nActions"]=rowTour+";"+nAction
        nPxP=dfTours.at[iTour,"PxP"]
        if nPxP=="":
            dfTours.at[iTour,"PxP"]=1
        else:
            dfTours.at[iTour,"PxP"]=nPxP+1       
```

```python
for index,row in dfTours[dfTours["nActions"]==""]["nTour Color".split(" ")].iterrows():
    dfTours.at[index,"Color"]=1
```

```python
with pd.ExcelWriter(nAssociations) as writer:
    dfActions.to_excel(writer, sheet_name=u'Actions',encoding="utf8",index=False)
    dfTours.to_excel(writer, sheet_name=u'Tours',encoding="utf8",index=False)

```
