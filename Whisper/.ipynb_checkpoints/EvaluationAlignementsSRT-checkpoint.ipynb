```python
import pandas as pd
import numpy as np
import codecs,re
```

# Lecture des sous-titres

```python
def lireSRT(nFile,lHeader=""):
    result={}
    with codecs.open(nFile,"r",encoding="utf8") as inFile:
        subSRT=[l.strip() for l in inFile.readlines()]
    
    print subSRT[:5]

    for l in subSRT:
        m=re.match(ur"^(\d+)$",l)
        if m:
            n=int(m.group(1))
            result[n]=[]
        else:
            m=re.match(ur"(.*) --> .*",l)
            if m:
                tDep=l.split(" --> ")[0]
                result[n].append(tDep)
                tFin=l.split(" --> ")[1]
                result[n].append(tFin)
            else:
                m=re.match(lHeader+ur"(.*)$",l)
                if m:
                    result[n].append(m.group(1))
    return result
```

```python
fManuelSRT="/Users/gilles/Downloads/FranceArgentine-FR-Comm.srt"
prefManuelSRT=ur"spk.-\d+\s+"
dSRT=lireSRT(fManuelSRT,prefManuelSRT)
dfManuel=pd.DataFrame.from_dict(dSRT,orient="index")
dfManuel.columns="start end com1 com2".split(" ")
dfManuel["startTime"]=pd.to_datetime(dfManuel["start"],format='%H:%M:%S,%f')
dfManuel["endTime"]=pd.to_datetime(dfManuel["end"],format='%H:%M:%S,%f')
dfManuel["start"]=dfManuel.startTime-dfManuel.iloc[0]["startTime"]
dfManuel["end"]=dfManuel.endTime-dfManuel.iloc[0]["startTime"]
dfManuel["longueur"]=(dfManuel.end-dfManuel.start).dt.total_seconds()
dfManuel
```

## Extraire une tranche de commentaires

```python
def extraireTours(deb,fin,df):
    mask=(df.start >= deb) & (df.start < fin)
    return df[mask]

display(extraireTours("0:0:0","0:1:0",dfManuel))
display(extraireTours("0:0:0","0:1:0",dfWhisper))
```
